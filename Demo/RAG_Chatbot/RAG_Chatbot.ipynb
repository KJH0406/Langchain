```python
# 라이브러리 설명
# 1. unstructured : 텍스트 파일 같은 구조화되지 않은 데이터 다루는데 사용
# 2. chromadb : 벡터를 저장 및 유사도 검색 지원
# 3. sentence_transformers: 문장을 벡터로 변환하고 이를 통해 텍스트 데이터의 의미적 유사성 계산
```

```python
# TextLoader 사용하여 텍스트 파일 로드
from langchain.document_loaders import TextLoader
documents = TextLoader("/Users/kimjangho/Desktop/Langchain_study/Data/AI.txt").load()
```

```python
# 큰 덩어리의 문서를 작은 덩어리로 분할
# 문장을 청크로 분할한다는 것은 자연어 처리(NLP)에서 사용하는 기술로, 문장을 의미있는 단위(청크)로 나누는 과정을 말함. 
# 이 청크는 단어 개별보다 큰 구성 요소로, 흔히 "구(phrase)"로 불림.주 목적은 문장 내에서 구문적으로 연관된 단어 그룹을 식별하고 구조화하는 것

from langchain.text_splitter import RecursiveCharacterTextSplitter

# 문서를 청크로 분할
def split_docs(documents,chunk_size=1000,chunk_overlap=20):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(documents)
  return docs

# docs 변수에 분할 문서를 저장
docs = split_docs(documents)
```

```python
# 벡터 데이터 베이스인 Chroma에 임베딩 처리된 벡터 저장

from langchain.embeddings import SentenceTransformerEmbeddings
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2") # 임베딩 모델 all-MiniLM-L6-v2 사용

# Chromdb에 벡터 저장
from langchain.vectorstores import Chroma
db = Chroma.from_documents(docs, embeddings)
```

```python
# 텍스트 파일에서 관련 내용을 찾아서 LLM에 제공하고, LLM은 해당 쿼리에 대한 답변 생성

import os #운영체제(os) 모듈을 가져올 때 사용되는 라이브러리
os.environ["OPENAI_API_KEY"] = "sk-" #openai 키 입력

from langchain.chat_models import ChatOpenAI
model_name = "gpt-3.5-turbo"  #GPT-3.5 turbo 모델 사용
# model_name = "gpt-4"  # 성능 비교를 위해서 GPT-4 모델 사용
llm = ChatOpenAI(model_name=model_name)

# Q&A 체인을 사용하여 쿼리에 대한 답변 얻기
from langchain.chains.question_answering import load_qa_chain
chain = load_qa_chain(llm, chain_type="stuff",verbose=True)

# 쿼리를 작성하고 유사성 검색을 수행하여 답변을 생성,따라서 txt에 있는 내용을 질의해야 함.
query = "AI란?"
matching_docs = db.similarity_search(query)
answer =  chain.run(input_documents=matching_docs, question=query)
answer
```
